```python
import csv

x=[]
y=[]
with open('iris.csv') as f:
    reader=csv.reader(f)
    next(reader)
    for row in reader:
        features=list(map(float,row[1:5]))
        label=row[5]
        x.append(features)
        y.append(label)
label_map={'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2}
y=[label_map[v] for v in y]
```

```python
petal_lengths=[x[i][0] for i in range(150)]
petal_widths=[x[i][1] for i in range(150)]
petal_lengths_0=[petal_lengths[i] for i in range(50)]
petal_widths_0=[petal_widths[i] for i in range(50)]
petal_lengths_1=[petal_lengths[i] for i in range(50,100)]
petal_widths_1=[petal_widths[i] for i in range(50,100)]
petal_lengths_2=[petal_lengths[i] for i in range(100,150)]
petal_widths_2=[petal_widths[i] for i in range(100,150)]

plt.scatter(petal_lengths_0,petal_widths_0,c='#f00')
plt.scatter(petal_lengths_1,petal_widths_1,c='#0f0')
plt.scatter(petal_lengths_2,petal_widths_2,c='#00f')
plt.show()
```

```python
sepal_lengths=[x[i][2] for i in range(150)]
sepal_widths=[x[i][3] for i in range(150)]
sepal_lengths_0=[sepal_lengths[i] for i in range(50)]
sepal_widths_0=[sepal_widths[i] for i in range(50)]
sepal_lengths_1=[sepal_lengths[i] for i in range(50,100)]
sepal_widths_1=[sepal_widths[i] for i in range(50,100)]
sepal_lengths_2=[sepal_lengths[i] for i in range(100,150)]
sepal_widths_2=[sepal_widths[i] for i in range(100,150)]

plt.scatter(sepal_lengths_0,sepal_widths_0,c='#f00')
plt.scatter(sepal_lengths_1,sepal_widths_1,c='#0f0')
plt.scatter(sepal_lengths_2,sepal_widths_2,c='#00f')
plt.show()
```
